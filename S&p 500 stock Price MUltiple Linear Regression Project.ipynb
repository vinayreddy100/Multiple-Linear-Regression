```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import dataset
```

```python
stock = pd.read_csv('S&P500_Stock_Data.csv')
```

```python
X = stock[['Interest Rates', 'Employment']]
```

```python
Y = stock[['S&P 500 Price']]
```

```python
X.shape
```

```python
Y.shape
```

```python
X
```

```python
Y
```

```python
### Train Model
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(fit_intercept = True)
```

```python
regressor.fit(X_train, Y_train)
```

```python
print('Linear Model coef(m)', regressor.coef_)
```

```python
print('Linear Model coef(b)', regressor.intercept_)
```

```python
## Evaluate Model
```

```python
y_predict = regressor.predict(X_test)
```

```python
y_predict
```

```python
Y_test
```

```python
# predict model_predictions to Y_test
```

```python
plt.scatter(Y_test, y_predict, color = 'black')
plt.ylabel('Model Predictions')
plt.xlabel('ytest real values')
plt.title('Multiple linear regression predictions')
```

```python
k = X_test.shape[1]
n = len(X_test)
```

```python
n
```

```python
k
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt
RMSE = float(format(np.sqrt(mean_squared_error(Y_test, y_predict)), '.3f'))
MSE = mean_squared_error(Y_test, y_predict)
MAE = mean_absolute_error(Y_test, y_predict)
r2 = r2_score(Y_test, y_predict)
adj_r2 = 1-(1-r2) * (n-1) / (n-k-1)
MAPE = np.mean(np.abs((Y_test - y_predict) / Y_test))* 100
```

```python
print('RMSE =' , RMSE, '\nMSE =', MSE, '\nMAE = ', MAE, '\r2 = ', r2, '\nadj_r2 = ', adj_r2, '\nMAPE = ', MAPE)
```
